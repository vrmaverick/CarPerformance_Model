```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sales = pd.read_csv('./data/sale_dataset.csv')
```

```python
sales.head()
```

```python
sales.size,sales.columns
```

```python
sales = sales.dropna()
```

```python
formated_sales= sales.rename(columns={"Price_in_thousands":"Price"})
formated_sales["Price"] = formated_sales["Price"]*1000
formated_sales["Total_profit_in thousands"]= formated_sales["Price"]*formated_sales["Sales_in_thousands"]

```

```python
formated_sales
```

```python
type(formated_sales["Price"][0]) #checking the Datatype
```

```python
formated_sales.loc[:, "cumulative_profit in millions"] = formated_sales["Total_profit_in thousands"].cumsum()/1000
```

```python
formated_sales
```

```python
formated_sales.sort_values('Latest_Launch',inplace=True)
```

# Data Visulaization

```python
fig,ax = plt.subplots()
ax.scatter(formated_sales["Power_perf_factor"],formated_sales["Sales_in_thousands"])
ax.set(xlabel="Power_Performance Factor",ylabel="Total_Sales(K)")
```

# using matplotlib Object oriented method

plot() Used for getting something quick but can hamper accuracy so use matplotlib

```python
s = formated_sales[formated_sales['Sales_in_thousands']>50] #filtering data
s.head()
```

```python
s.head()
```

```python
fig, ax = plt.subplots(figsize=(30,10))
s.plot(kind="scatter",
                    x="Price",y="Fuel_efficiency",c="Power_perf_factor",ax=ax);
ax.set_xlim([10000,38000])
```

```python
fig, ax = plt.subplots(figsize=(30,10))
s.plot(kind="scatter",
                    x="Price",y="Fuel_efficiency",c="Engine_size",ax=ax);
ax.set_xlim([10000,38000]);
```
